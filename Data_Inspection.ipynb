Import libraries

```python
import pandas as pd
import numpy as np

from pathlib import Path

import warnings
warnings.filterwarnings("ignore")

# Display options
pd.set_option("display.max_columns", None)
pd.set_option("display.max_rows", 100)
```

Set dataset and output paths

```python
# Google Drive location

PROJECT_DIR = Path("/content/drive/MyDrive/MSc_Dissertation")

DATASET_DIR = PROJECT_DIR / "Dataset"

OUTPUT_DIR = DATASET_DIR / "Processed"

OUTPUT_DIR.mkdir(exist_ok=True)

DATA_FILE = DATASET_DIR / "Friday-16-02-2018_TrafficForML_CICFlowMeter.csv"
```

Load the dataset

```python
df = pd.read_csv(DATA_FILE, low_memory=False)

print("Dataset loaded successfully.")
```

Initial dataset inspection

```python
print("="*60)

print("Dataset Shape")
print(df.shape)

print("="*60)

print("Column Count")
print(len(df.columns))

print("="*60)

print("Data Types")
print(df.dtypes.value_counts())

print("="*60)

print("Memory Usage")

memory = df.memory_usage(deep=True).sum() / 1024**2

print(f"{memory:.2f} MB")
```

Preview dataset and information

```python
df.head()
df.info()
```

Class distribution

```python
print(df["Label"].value_counts())

print()

print(df["Label"].value_counts(normalize=True) * 100)
```

Missing values

```python
missing = df.isnull().sum()

missing = missing[missing > 0]

print("Missing columns:")

display(missing)
```

Duplicate values

```python
duplicates = df.duplicated().sum()

print(f"Duplicate rows: {duplicates}")
```

Infinite values

```python
numeric_df = df.select_dtypes(include=[np.number])

print("Infinite values:")

print(np.isinf(numeric_df).sum().sum())
```

Summary of the inspection

```python
df.describe(include="all")
```

```python
summary = {
    "Rows": len(df),
    "Columns": len(df.columns),
    "Duplicates": duplicates,
    "Missing Values": int(df.isnull().sum().sum()),
    "Memory(MB)": round(memory,2)
}

summary
```
